Berikut merupakan implementasi program Python yang digunakan untuk membentuk matriks transisi dan menghitung distribusi stasioner pada model rantai Markov tingkat kehadiran mahasiswa.

```python
import numpy as np

# ==========================================
# STATE
# ==========================================

states = ["Rajin Hadir", "Kadang Hadir", "Sering Absen"]

# ==========================================
# INPUT MATRIKS TRANSISI 3x3
# ==========================================

P = []

print("Masukkan probabilitas transisi:")

for i in range(3):

    print(f"\nDari state {states[i]}:")

    baris = []
    total = 0

    for j in range(3):
        p = float(input(f"Peluang ke {states[j]}: "))
        baris.append(p)
        total += p

    # Validasi jumlah probabilitas
    if abs(total - 1) > 0.0001:
        print("Jumlah probabilitas dalam satu baris harus = 1")
        exit()

    P.append(baris)

# Ubah menjadi array numpy
P = np.array(P)

# ==========================================
# TAMPILKAN MATRIKS
# ==========================================

print("\nMatriks Transisi:")
print(P)

# ==========================================
# HITUNG STEADY STATE
# ==========================================

A = P.T - np.eye(3)

# Tambah syarat jumlah probabilitas = 1
A = np.vstack((A, np.ones(3)))

b = np.array([0, 0, 0, 1])

# Penyelesaian SPL
steady_state = np.linalg.lstsq(A, b, rcond=None)[0]

# ==========================================
# OUTPUT
# ==========================================

print("\nDistribusi Stasioner:")

for i in range(3):
    print(f"{states[i]} = {steady_state[i]:.4f}")
```
